```python
from ultralytics import YOLO
import cv2

#load a pre trained Yolo modedl
model = YOLO("yolov8n.pt")    # have used 8n as it is fast and lightweight for more accuracy 8s is prefered.

#webcam
capture = cv2.VideoCapture(2)  # use 0 if you donot have any other mac environemnt around you like iphone and ipad.

while True:
    ret , frame = capture.read()
    if not ret:
        break
    
    results = model(frame) # run frame on yolo
    
    annotated_frame = results[0].plot() 
    
    cv2.imshow("YOLOv8 Detection", annotated_frame) # display frame with bounding box
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
    
# release resources
capture.release() 
cv2.destroyAllWindows()
    
```
